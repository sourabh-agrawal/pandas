## prime no problem

```python
import math
count = 1    #since 2 is also prime no
print("{} is a prime no".format(2))
i = 2
while(count<100):
    i += 1
    flag = 0
    for j in range(2, math.ceil(math.sqrt(i))+1):
        if i%j == 0:
            flag = 1
            break
        else:
            flag = 0
        
    if flag == 0:
        print("{} is a prime no".format(i))
        count += 1
        
print(count)
```

## car rental system

```python
import sys
'''
Section 1: Collect customer input
'''
# print("(B)udget, (D)aily, or (W)eekly rental?\n")
rentalCode = input ("(B)udget, (D)aily, or (W)eekly rental?\n").upper()

 
#a loop is provided here to check if user inputs are in a valid format
#if not in valid format, the code will ask the user to re-enter information
while rentalCode not in ("W", "D", "B", "w", "d", "b"):  #repeating loop until input is satisfied.
  print ("Rental code not valid.  Please re-enter the rental code in W, D, or B format.  Thank You.")  #reminds user of correct input
  rentalCode = input ("(B)udget, (D)aily, or (W)eekly rental?\n").upper()  #restats question
 
 

rentalPeriod=0
#this is an if-else statement to resolve the time of rental based on rental type
if rentalCode == 'B' or rentalCode == 'D':  #budget & daily go by number of days
    rentalPeriod = int(input ("Number of Days Rented:\n"))
else:
    rentalPeriod = int(input ("Number of Weeks Rented:\n"))  #for weekly rentals
# print(rentalCode)
# print(rentalPeriod)



budgetCharge = 40.00
dailyCharge = 60.00
weeklyCharge = 190.00 
 
baseCharge = 0   #defining helper variable

if rentalCode == 'B':  #if statement for Budget rentals
  baseCharge = int(rentalPeriod) * int(budgetCharge)  #math baseCharge for Budget rentals
elif rentalCode == 'D':  #if statement for Daily rentals
  baseCharge = int(rentalPeriod) * int(dailyCharge)  #math baseCharge for Daily rentals
elif rentalCode == 'W':  #if statement for Weekly rentals
  baseCharge = int(rentalPeriod) * int(weeklyCharge)  #math baseCharge for Weekly rentals

 
odoStart = input("Starting Odometer Reading:\n")  #prompts user for starting odometer information
 
odoEnd = input("Ending Odometer Reading:\n")  #prompts user for ending odometer information
 
# print(odoStart)
# print(odoEnd)
# print("%.1f" %float(baseCharge))
 
'''
Section 2: Calculate the costs from the customer input
'''
 
totalMiles = int(odoEnd) - int(odoStart)  #math to calculate totalMiles  #math to calculate totalMiles
# print(totalMiles)

mileCharge = 0

if rentalCode == 'B':
  mileCharge = totalMiles*(0.25)
elif rentalCode =='D':
  averageDayMiles = totalMiles/rentalPeriod
  extraMiles = 0
  if averageDayMiles <= 100:
    extraMiles = 0
  else:
    extraMiles = averageDayMiles - 100
  mileCharge = (0.25)*extraMiles*rentalPeriod
elif rentalCode=='W':
  averageWeekMiles = totalMiles/rentalPeriod
  if averageWeekMiles>900:
    mileCharge = 100*rentalPeriod
  else:
    mileCharge = 0

# print(mileCharge)

amtDue = baseCharge + mileCharge

 
'''
Section 3: Display the results to the customer
'''

print ("Rental Summary")
print ("Rental Code: {}".format(rentalCode))
print ("Rental Period: {}".format(rentalPeriod))
print ("Starting Odometer: {}".format(odoStart))
print("Ending Odometer: {}".format(odoEnd))
print ("Miles Driven: {}".format(totalMiles))
print ("Amount Due: $%.2f" %amtDue)
```
